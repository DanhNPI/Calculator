```python
# Thiết kế menu
from tkinter import *

food = ['Bún bò', 'Bún chả', 'Bún đậu', 'Trà đá']
price = [40000, 55000, 50000, 10000]
menu = {}
for index in range(len(food)):
    menu[food[index]] = price[index]
print (menu['Bún bò'])

def update():
    selBB = b_bo.get()
    selBC = b_cha.get()
    selBD = b_dau.get()
    selTD = t_da.get()

    cosBB = int(selBB) * int(entryBB.get())
    cosBC = int(selBC) * int(entryBC.get())
    cosBD = int(selBD) * int(entryBD.get())
    cosTD = int(selTD) * int(entryTD.get())   

    lblBB['text'] = cosBB
    lblBC['text'] = cosBC 
    lblBD['text'] = cosBD
    lblTD['text'] = cosTD

    total = cosBB + cosBC + cosBD + cosTD

    if choice.get() == 'Tiền mặt':
        total = total - (total*0.05)
    else:
        total = total + (total*0.05)

    result['text'] = str(total) + 'VND'

root = Tk()
b_bo = StringVar()
b_cha = StringVar()
b_dau = StringVar()
t_da = StringVar()
x = StringVar()
x1 = StringVar()
x2 = StringVar()
x3 = StringVar()
choice = StringVar()
x.set('0')
x1.set('0')
x2.set('0')
x3.set('0')

b_bo.set(0)
b_cha.set(0)
b_dau.set(0)
t_da.set(0)

choice.set('Tiền mặt')
frame = LabelFrame( root, 
                    text='Menu',
                    font=('Arial', 13),
                    width=400,
                    height=300, labelanchor="n",
                    fg = 'red')
frame.pack(anchor='w')
'--------------------------------------------------- Tên món ----------------------------------------------------------------------------'
name_food = Label(frame, text='Tên món', font=('Calibri', 15, 'bold'), relief=RAISED, fg='red', bg='yellow')
name_food.grid(column=0, row=0, padx=5)

btnBB = Checkbutton(frame, text=food[0], font=('Arial', 13), variable= b_bo, onvalue=price[0], offvalue=0, command=update)
btnBB.grid(column=0, row=1, sticky='w')

btnBC = Checkbutton(frame, text=food[1], font=('Arial', 13), variable= b_cha, onvalue=price[1], offvalue=0, command=update)
btnBC.grid(column=0, row=2, sticky='w')

btnBD = Checkbutton(frame, text=food[2], font=('Arial', 13), variable= b_dau, onvalue=price[2], offvalue=0, command=update)
btnBD.grid(column=0, row=3, sticky='w')

btnTD = Checkbutton(frame, text=food[3], font=('Arial', 13), variable= t_da, onvalue=price[3], offvalue=0, command=update)
btnTD.grid(column=0, row=4, sticky='w')
'----------------------------------------------------- Giá 1 bát-----------------------------------------------------------------------'
money = Label(frame, text='Giá 1 bát', font=('Calibri', 15, 'bold'), relief=RAISED, fg='red', bg='yellow')
money.grid(column=1, row=0, padx=40)

giaBB = Label(frame, text=str(price[0]), font=('Arial', 13))
giaBB.grid(column=1, row=1, padx=70, sticky='w')

giaBC = Label(frame, text=str(price[1]), font=('Arial', 13))
giaBC.grid(column=1, row=2, sticky='w', padx=70)

giaBD = Label(frame, text=str(price[2]), font=('Arial', 13))
giaBD.grid(column=1, row=3, sticky='w', padx=70)

giaTD = Label(frame, text=str(price[3]), font=('Arial', 13))
giaTD.grid(column=1, row=4, sticky='w', padx=70)
'-------------------------------------------------------------------------------------------------------------------------------'
s_luong = Label(frame, text='Số lượng', font=('Calibri', 15, 'bold'), relief=RAISED, fg='red', bg='yellow')
s_luong.grid(column=2, row=0, padx=10)

entryBB = Entry(frame, font=('Arial', 13), bd=3, relief=RAISED, width=9, justify='center', textvariable = x)
entryBB.grid(column=2, row=1, padx=10, sticky='w')

entryBC = Entry(frame, font=('Arial', 13), bd=3, relief=RAISED, width=9, justify='center', textvariable = x1)
entryBC.grid(column=2, row=2, sticky='w', padx=10)

entryBD = Entry(frame, font=('Arial', 13), bd=3, relief=RAISED, width=9, justify='center', textvariable = x2)
entryBD.grid(column=2, row=3, sticky='w', padx=10)

entryTD = Entry(frame, font=('Arial', 13), bd=3, relief=RAISED, width=9, justify='center', textvariable = x3)
entryTD.grid(column=2, row=4, sticky='w', padx=10)
'-------------------------------------------------------------------------------------------------------------------------------'
s_luong = Label(frame, text='Thành tiền', font=('Calibri', 15, 'bold'), relief=RAISED, fg='red', bg='yellow')
s_luong.grid(column=3, row=0, padx=60)

lblBB = Label(frame, text='', font=('Arial', 13))
lblBB.grid(column=3, row=1)

lblBC = Label(frame, text='', font=('Arial', 13))
lblBC.grid(column=3, row=2)

lblBD = Label(frame, text='', font=('Arial', 13))
lblBD.grid(column=3, row=3)

lblTD = Label(frame, text='', font=('Arial', 13))
lblTD.grid(column=3, row=4)
'-------------------------------------------------------------------------------------------------------------------------------'
frame1 = LabelFrame( root, 
                    text='Phương thức',
                    font=('Arial', 13),
                    width=400,
                    height=20, labelanchor='n'
                    fg = 'red')
frame1.pack(anchor='w', pady=20)

rd1 = Radiobutton(frame1, text='Tiền mặt -5%', font=('Arial', 13), variable=choice, value='Tiền mặt')
rd1.grid(column=0, row=0, pady=20, padx=10)
rd2 = Radiobutton(frame1, text='Chuyển khoản +5%', font=('Arial', 13), variable=choice, value='Chuyển khoản')
rd2.grid(column=1, row=0, pady=20, padx=20)
'-------------------------------------------------------------------------------------------------------------------------------'
frame2 = LabelFrame( root, 
                    text='Tổng tiền',
                    font=('Arial', 13),
                    width=200,
                    height=200, labelanchor='n'
                    fg = 'red', bd=5, bg='Black',
                    highlightthickness=5)
frame2.pack(side=LEFT, pady=10)

result = Label(frame2, text =" ", font=('Arial', 13), fg = 'red', bg = 'yellow')
result.pack(side=LEFT, padx=30)
'-------------------------------------------------------------------------------------------------------------------------------'

btnUpdate = Button(root, text='Update', height=4, width=9, font=('Arial', 10), bd=3, activeforeground='red', activebackground='green', command=update)
btnUpdate.place(x =400, y = 202)

root.mainloop()

```
